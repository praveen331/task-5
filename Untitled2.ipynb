```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the CSV file
df=pd.read_csv('https://raw.githubusercontent.com/kislay960/The-Sparks-Foundation/master/SampleSuperstore.csv')
df.head()
```

```python
print(df.shape)
```

```python
df.describe(include="all")
```

```python
print(df.isna().sum())
```

```python
df.drop_duplicates(keep = "first",inplace = True)
```

```python
print(df.corr())
```

```python
df.groupby('Ship Mode').mean()
```

```python
df.groupby('Segment').mean()
```

```python
corr = df.corr()
sns.heatmap(corr,annot=True)
```

```python
sns.countplot(df['Segment'])
plt.show()
```

```python
sns.countplot(df['Segment'])
plt.show()
```

```python
sns.countplot(df['State'])
plt.show()
```

```python
plt.plot(df['Sales'])
plt.show()
```

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
sns.pairplot(df,hue='Segment');
```

```python
plt.figure(figsize=(10,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,8))
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
data1 = df[['City','Category', 'State','Sales','Profit']]
# pandas pivot with multiple variables
heatmap1_df= pd.pivot_table(data1,values='Sales', index=["City"], columns='Category')
```

```python
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap1_df, cmap="BuGn")
```

```python
heatmap2_data = pd.pivot_table(data1,values='Sales', index=["State"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap2_data, cmap="BuGn")
```

```python
heatmap3_data = pd.pivot_table(data1,values='Profit', index=["State"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap3_data, cmap="BuGn")
```

```python
heatmap4_data = pd.pivot_table(data1,values='Profit', index=["City"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap4_data, cmap="BuGn")
```

```python
heatmap5_df = pd.pivot_table(df,values='Profit', index=["City"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap5_df, cmap="BuGn")
```

```python
heatmap6_df = pd.pivot_table(df,values='Profit', index=["State"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap6_df, cmap="BuGn")
```

```python
plt.figure(figsize=(11,11))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```
